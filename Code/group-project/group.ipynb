```python
import pandas as pd
```

```python
sections = ["Sec 001-801", "Sec 003-803", "Sec 004-804"]

dfs = []
for section in sections:
    dfs.append(
        {"section": section, "df": pd.read_excel("Group.xlsx", sheet_name=section)}
    )
```

```python
for (idx, item) in enumerate(dfs):
    df1 = item["df"].iloc[2:, :].copy()
    colsOri = df1.columns.values
    cols = ["group", "group_name", *colsOri[2:]]
    df1.columns = cols
    df1 = df1.reset_index(drop=True)
    dfs[idx]["df1"] = df1

```

```python
data = []

def process(row, section):
    rowDrop = row.iloc[2:].dropna()

    if rowDrop.shape[0] == 2:
        return None

    group = row.loc["group"]
    group_name = row.loc["group_name"]

    for (idx, item) in enumerate(rowDrop):
        if idx % 2 == 0:
            print(item)
            data.append(
                {
                    "section": section,
                    "group": group,
                    "group_name": group_name,
                    "name": item,
                    "id": rowDrop.iloc[idx + 1],
                }
            )


for item in dfs:
    df1 = item["df1"]
    section = item["section"]
    df1.apply(lambda row: process(row, section), axis=1)
```

```python
dfAll = pd.DataFrame(data)
display(dfAll)
```
