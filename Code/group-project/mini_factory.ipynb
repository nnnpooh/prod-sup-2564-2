```python
df4 = df3[~df3["sec"].isna()]
df4
```

```python
df5 = df4.append(dffill, ignore_index=True)
df5 = df5.sort_values(by=["lecture", "group", "cmu_id"])
```

```python
df5
```

```python
lectures = ["Sec 001-801", "Sec 003-803", "Sec 004-804"]
days = [
    {"day1": "14 กพ 8.00-9.30", "day2": "17 กพ 8.00-9.30"},
    {"day1": "14 กพ 11.00-12.30", "day2": "17 กพ 11.00-12.30"},
    {"day1": "15 กพ 8.00-9.30", "day2": "18 กพ 8.00-9.30"},
]
dfAll = df5
data = []
for (lecture, days) in zip(lectures,days):
    dfsec = dfAll[dfAll["lecture"] == lecture].reset_index(drop=True)
    dfsec = dfsec[["cmu_id", "name", "sec", "lecture"]]
    dfsec.loc[0:23, "day"] = 1
    dfsec.loc[0:23, "datetime"] = days['day1']
    dfsec.loc[24:, "day"] = 2
    dfsec.loc[24:, "datetime"] = days['day2']
    data.append(dfsec.copy())


# dfs1 = df5[df5['lecture'] == 'Sec 001-801'].reset_index(drop=True)
# dfs3 = df5[df5['lecture'] == 'Sec 003-803'].reset_index(drop=True)
# dfs4 = df5[df5['lecture'] == 'Sec 004-804'].reset_index(drop=True)
```

```python
with pd.ExcelWriter("MiniFac_processed.xlsx") as writer:
    data[0].to_excel(writer, sheet_name="Sec 001-801", index=False)
    data[1].to_excel(writer, sheet_name="Sec 003-803", index=False)
    data[2].to_excel(writer, sheet_name="Sec 004-804", index=False)

```
